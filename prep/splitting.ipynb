```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('no_pert.csv')
```

```python
np.unique(df.ident)
```

```python
ids=[str(i) for i in np.unique(df.ident)]
```

```python
idx=[(4-len(i))*'0'+i for i in ids]
```

```python
idx
```

```python
for i,lab in zip(ids,idx):
    filter=df.ident==int(i)
    df[filter].to_csv('results'+lab+'.txt')
```
